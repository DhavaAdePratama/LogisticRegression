```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/DhavaAdePratama/LogisticRegression/refs/heads/main/SMSSpamCollection',
                  sep = '\t',
                 header = None,
                 names = ['label', 'message']
              )
df.head()
```

```python
#Menghitung berapa banyak data pada label ham dan spam

df['label'].value_counts()
```

```python
# preprocessing
from sklearn.preprocessing import LabelBinarizer

x = df['message']
y = df['label']

lb = LabelBinarizer()
y = lb.fit_transform(y).ravel() # hasil dari lb.fit transform yang dikenakan pada variable y akan menghasilkan array multidimensi sehingga perlu diubah menjadi satu dimensi menggunakan ravel()

lb.classes_ #melihat proses label binarizer
```

```python
#pembagian nilai training dan testing

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

print(x_train, '\n')
print(y_train)
```

```python
#Feature Extraction dengan TF-IDF
#menggunakan TF-IDF karena sebuah mesin tidak dapat menyerap data kategorik

from sklearn.feature_extraction.text import TfidfVectorizer

Vectorizer = TfidfVectorizer(stop_words = 'english')

x_train_tfidf = Vectorizer.fit_transform(x_train)
x_test_tfidf = Vectorizer.transform(x_test)

print(x_train_tfidf)
```

```python
# Binary Classification dengan Logistic Regression

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train_tfidf, y_train)
y_pred = model.predict(x_test_tfidf)

for pred, sms in zip(y_pred[:10], x_test[:10]):
    print(f'PREDICT : {pred} - SMS {sms} ')
```
